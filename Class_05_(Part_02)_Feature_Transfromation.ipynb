# Robust Scaler

```python
from sklearn.preprocessing import RobustScaler
rbs = RobustScaler()
df4 = df.copy()
```

```python
for column in df4.columns:
    if column != "Profit":
      if pd.to_numeric(df4[column], errors='coerce').notnull().all():
          df4[ [column]] = rbs.fit_transform(df4[[column]])
```
